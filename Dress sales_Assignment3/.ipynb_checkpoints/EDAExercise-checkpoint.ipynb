You can see that there are two types of variables one with a large number of missing values and another is very less number of missing values. These two columns can be categorized as:

Type-1: Missing values are very less (around 2 or 3 missing values): Price, Season, NeckLine, SleeveLength, Winter and Autumn. 

Type-2: Missing values are large in numbers (more than 15%): Material, FabricType, Decoration and Pattern Type.



```python
inp0.describe()
```

```python
# Deal with the missing values of Type-1 columns: Price, Season, NeckLine, SleeveLength, Winter and Autumn.

inp0["Price"].fillna(inp0["Price"].mode()[0], inplace=True)
inp0["Season"].fillna(inp0["Season"].mode()[0], inplace=True)
inp0["NeckLine"].fillna(inp0["NeckLine"].mode()[0], inplace=True)
inp0["SleeveLength"].fillna(inp0["SleeveLength"].mode()[0], inplace=True)
inp0["Autumn"].fillna(inp0["Autumn"].median(), inplace=True)
inp0["Winter"].fillna(inp0["Winter"].median(), inplace=True)
```

```python
# Deal with the missing values for Type-2 columns: Material,FabricType , Decoration and Pattern Type.
inp0["Material"].fillna(inp0["Material"].mode()[0], inplace=True)
inp0["FabricType"].fillna(inp0["FabricType"].mode()[0], inplace=True)
inp0["Decoration"].fillna(inp0["Decoration"].mode()[0], inplace=True)
inp0["Pattern Type"].fillna(inp0["Pattern Type"].mode()[0], inplace=True)
```

```python
temp = inp0_sum.copy()
```

```python
def df_column_normalize(dataframe, percent=False):
   if percent:
        return dataframe.div(dataframe.sum(axis=0), axis=1)*100
   else:
        return dataframe.div(dataframe.sum(axis=0), axis=1)
```

```python
temp = df_column_normalize(temp, percent=True)
```

```python
fig4, ax = plt.subplots(figsize=(30,20))
fig4.subplots_adjust(top=.94)
plt.suptitle('“Style” and “Price” category vs recommendations', fontsize=20, fontweight='bold')


sns.heatmap(temp, annot=True, fmt='.2f', linewidths=.3, ax=ax, cmap='RdPu');
```
